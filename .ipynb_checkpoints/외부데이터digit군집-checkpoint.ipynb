```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from minisom import MiniSom
from pylab import plot, axis, show, pcolor, colorbar, bone
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
csvfile = "C:/Users/sylee/Documents/GitHub/DataSafe/DataSafe/92dataNS.csv"
df = pd.read_csv(csvfile)

df.head(3)
```

```python
df['지역'] = df['광역시도']+df['시군구명']
df.drop(columns=['광역시도','시군구명','총면적'],axis=1, inplace=True)
```

```python
col1=df.columns[-1:].to_list()
col2=df.columns[:-1].to_list()
new_col=col1+col2
dataset=df[new_col]
dataset.head(3)
```

```python
X = dataset.iloc[:, 1:-3]
X
```

```python
y = dataset.iloc[:,-1:]
y
```

```python
data = X.to_numpy()
data
```

```python
plt.plot(data) #checking the scaling result of A2
```

```python
labels = y
y
```

```python
# labels
(labels['cluster']==0).sum()
```

```python
# Training the SOM
#from minisom import MiniSom
som = MiniSom(x = 16, y = 16, input_len = 33, sigma = 0.5, learning_rate = 0.7)
som.random_weights_init(data)
som.train_random(data = data, num_iteration = 100)
# Visualizing the results
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
```

```python
# ## labels -> k-meads 군집 라벨링
# labels['cluster'] = 0 
# labels['cluster'] = 1 
# labels['cluster'] = 2 
# labels['cluster'] = 3 
# labels['cluster'] = 4 
# labels['cluster'] = 5 
# labels['cluster'] = 6 
# labels['cluster'] = 7 
# labels['cluster'] = 8 
# # labels[labels=='9'] = 9 

markers = ['o', 'v', '1', '3', '8', 's', 'p', 'x', 'D']
colors = ["r", "g", "b", "y", "c", (0,0.1,0.8), (1,0.5,0), (1,1,0.3), "m"]

for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```
