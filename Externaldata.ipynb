```python
import numpy as np
df4['광역시도'] = np.where(df4['광역시도']=='서울', '서울특별시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='인천', '인천광역시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='부산', '부산광역시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='대전', '대전광역시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='대구', '대구광역시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='광주', '광주광역시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='울산', '울산광역시',df4['광역시도'])
df4['광역시도'] = np.where(df4['광역시도']=='강원', '강원도',df4['광역시도'])
```
